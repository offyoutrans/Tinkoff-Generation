```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.random.choice
```

```python
p = np.linspace(0, .75, 1000)
exp_mean = np.zeros(shape=p.size)
values = np.array([0, 1])
```

```python
def attm(p):
    c = np.array([1])
    el = 1
    while el:
        el=r(a = values, size = 1, p = np.array([1-p, p]))
        c+=1
    return c
```

```python
def experiment(n):
    results = np.zeros(shape=p.size)
    data = np.zeros(shape=1000)
    for i in range(len(p)):
        for j in range(n):
            data[j] = attm(p[i])
        results[i] = data.mean()
    return results
```

```python
def hypothesis(p):
    return 1/(1-p)
```

```python
observation = experiment(1000)
```

```python
expectation = hypothesis(p)
```

```python
plt.plot(p, observation)
```

```python
plt.plot(p, expectation)
```

```python
plt.plot(p, (observation-expectation)**2/2)
```
