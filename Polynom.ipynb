```python
#import ipywidgets as widgets
#from ipywidgets import interact, interact_manual
#import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def f(x, y):
    return x**4+2*x**2*y+10*x**2+6*x*y+10*y**2-6*x+4
```

```python
xs = np.linspace(-3, 3, 30)
ys = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(xs, ys)
Z = f(X, Y)
plane = np.zeros((30, 30))+f(1/3,-1/9)
```

```python
fig = plt.figure()
#plt.hold(True)
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, color = 'green')
ax.plot_surface(X, Y, plane, color='black')
ax.scatter(1/3, -1/9, f(1/3, -1/9), color='red', marker = 'o', linewidth=3)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('f(x, y)')
ax.view_init(210, 83)
#for angle in range(0, 360):
#    ax.view_init(30, angle)
#    plt.draw()
#    plt.pause(.001)
```
